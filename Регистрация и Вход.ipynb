```python
from tkinter import *
import tkinter.ttk as ttk
window = Tk()
def pro_reg():
    global poch
    global reg
    global par
    global combobox
    global nom
    global pr_12
    global nom1
    nom=2
    nom1=0
    if nom1==12:
        pr_12.withdraw()
    reg=Toplevel()
    poch = StringVar()
    par= StringVar()
    reg.title("Регистрация")
    lb1 = Label(reg, text="Логин", font=("Times New Roman", 14))  
    poch_entry = Entry(reg,textvariable=poch)
    poch_entry.grid(row=0,column=1, padx=5, pady=5)
    lb1.grid(column=0, row=0)
    lb2 = Label(reg, text="Пароль", font=("Times New Roman", 14))  
    par_entry = Entry(reg,textvariable=par)
    par_entry.grid(row=1,column=1, padx=5, pady=5)
    lb2.grid(column=0, row=1)
    lb3 = Label(reg, text="Группа", font=("Times New Roman", 14)) 
    lb3.grid(column=0, row=2)
    combobox = ttk.Combobox(reg ,values =[u"1",u"2",u"3",u"4",u"5",u"6"],height=6)
    combobox.grid(column=1,row=2)
    btn = Button(reg, text="Регистрация", command=pr, bg="#ffffe0")
    btn.grid(column=0, row=3)
    btn1 = Button(reg, text="Вход", command=com, bg="#ffffe0")
    btn1.grid(column=1, row=3)
    reg.mainloop()
def reg():
    window.withdraw()
    pro_reg()
def reg_com():
    global com
    com.withdraw()
    pro_reg()
def pof_reg():
    global pr_1
    pr_1.withdraw()
    pro_reg()
def pr_1():
    global pr_1
    pr_1=Toplevel()
    pr_1.title("Ошибка")
    lb1 = Label(pr_1, text="Что-то не так. Попробуйте снова", font=("Times New Roman", 14))
    lb1.grid(column=0, row=0)
    btn = Button(pr_1, text="Регистрация. Повтор", command=pof_reg, bg="#ffffe0")
    btn.grid(column=0, row=1)
    pr_1.mainloop()
def pr_2():
    global combobox
    global pr
    pr=Toplevel()
    pr.title("Продолжение")
    btn = Button(pr, text="Вход", command=com, bg="#ffffe0")
    btn.grid(column=0, row=1)
    lb1 = Label(pr, text="Данные введены корректно.", font=("Times New Roman", 14))
    lb1.grid(column=0, row=0)
    pr.mainloop()
def pr():
    global poch
    global par
    global combobox
    global reg
    reg.withdraw()
    em_pas=[]
    with open('email.txt', "r") as f:
        for line in f:
            a11=line.split(";")
            em_pas.append(a11)
    ooo=0
    for i in em_pas:
        if poch.get()==i[0]:
            ooo=1
    if ((poch.get()=="") or (par.get()=="") or (combobox.get()=="")):
        pr_1()
    if poch.get()!="" and par.get()!="" and combobox.get()!="" and ooo!=1:
        global nom
        nom=3
        with open('email.txt', "a") as f:
            zapis=poch.get()+";"+par.get()+";"+combobox.get()+"\n"
            f.write(zapis)
        pr_2()
    if ooo==1:
        pr_1()
def com():
    global nom
    global com
    global poch
    global par
    if nom==1:
       window.withdraw()
    if nom==2:
       reg.withdraw()
    if nom==3:
       pr.withdraw()
    if nom==4:
       pr_12.withdraw()
    com=Toplevel()
    poch = StringVar()
    par= StringVar()
    com.title("Вход")
    lb1 = Label(com, text="Логин", font=("Times New Roman", 14))  
    poch_entry = Entry(com,textvariable=poch)
    poch_entry.grid(row=0,column=1, padx=5, pady=5)
    lb1.grid(column=0, row=0)
    lb2 = Label(com, text="Пароль", font=("Times New Roman", 14))  
    par_entry = Entry(com,textvariable=par)
    par_entry.grid(row=1,column=1, padx=5, pady=5)
    lb2.grid(column=0, row=1)
    btn = Button(com, text="Регистрация", command=reg_com, bg="#ffffe0")
    btn.grid(column=1, row=3)
    btn1 = Button(com, text="Вход", command=itog, bg="#ffffe0")
    btn1.grid(column=0, row=3)
    com.mainloop()
def itog():
    global poch
    global par
    em_pas=[]
    with open('email.txt', "r") as f:
        for line in f:
            a11=line.split(";")
            em_pas.append(a11)
    ooo=0
    for i in em_pas:
        if poch.get()==i[0] and par.get()==i[1]:
            ooo=1
    if ooo==1:
        global com
        com.withdraw()
        itog_21()
    else:
        com.withdraw()
        itog_12()
def itog_12():
    global pr_12
    global nom1
    nom1=12
    pr_12=Toplevel()
    pr_12.title("Ошибка")
    lb12 = Label(pr_12, text="Такого аккаунта нет", font=("Times New Roman", 14))
    lb12.grid(column=0, row=0)
    btn1 = Button(pr_12, text="Регистрация", command=pro_reg, bg="#ffffe0")
    btn1.grid(column=0, row=1)
    pr_12.mainloop()
import webbrowser
def callback(event):
    webbrowser.open_new(event.widget.cget("text"))
def itog_21():
    pr_21=Toplevel()
    pr_21.title("Вход выполнен")
    frame1=Frame(pr_21,bg='green',bd=100)
    lbl = Label(frame1, text=r"http://www.fa.ru/Pages/Home.aspx", fg="blue", cursor="hand2")
    lbl.pack()
    lbl.bind("<Button-1>", callback)
    frame1.pack()
    pr_21.mainloop()
def nach():
    global nom
    nom=1
    window.title("Авторизация")
    lbl = Label(window, text="Регистрация", font=("Times New Roman", 14))  
    lbl.grid(column=0, row=0)  
    btn = Button(window, text="Перейти", command=reg)  
    btn.grid(column=1, row=0)
    lbl = Label(window, text="Вход", font=("Times New Roman", 14))  
    lbl.grid(column=0, row=1)  
    btn = Button(window, text="Перейти", command=com)  
    btn.grid(column=1, row=1)
    window.mainloop()
nach()
```
